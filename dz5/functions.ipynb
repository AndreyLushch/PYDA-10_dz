# Домашнее задание к лекции "Функции"

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

1. перечень всех документов
```
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```
2. перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
```
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Общие требования к программе:
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **documents** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

````python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}


def name_by_doc():  # - узнать владельца документа по его номеру (p)
    number_doc = input('Введите номер документа: ')
    flag = False
    for i in documents:
        if i['number'] == number_doc:
            print('Владелец документа:', i['name'])
            flag = True
            break
    if flag == False:
        print('Документ не найден в базе')
    

def dir_by_doc():  # - по номеру документа узнать на какой полке он храниться (s)
    number_doc = input('Введите номер документа: ')
    flag = False
    num = len(list(directories.keys()))
    for i in range(1, num + 1):
        i = str(i)
        if number_doc in directories[i]:
            print('Документ хранится на полке:', i)
            flag = True
            break
    if flag == False:
        print('Документ не найден в базе')
        

def all_doc():  # - вывести полную информацию по всем документам (l)
    num = len(list(directories.keys()))
    for d in documents:
        for i in range(1, num + 1):
            i = str(i)
            if d['number'] in directories[i]:
                print('№ ', d['number'], ', тип: ', d['type'], ', владелец: ', d['name'], ', полка хранения: ', i, sep='')
                break


def add_dir():  # - добавить новую папку (as)
    number_new = input('Введите номер полки: ')
    if number_new not in list(directories.keys()):
        directories[number_new] = []
        print('Полка добавлена. Текущий перечень полок: ', end='')
        print(*list(directories.keys()), sep=', ')
    else:
        print('Такая полка уже существует. Текущий перечень полок: ', end='')
        print(*list(directories.keys()), sep=', ')


def del_dir():  # - удалить существующую папку при условии, что она пустая (ds)
    number_del = input('Введите номер полки: ')
    if number_del in list(directories.keys()) and directories[number_del] == []:
        del directories[number_del]
        print('Полка удалена. Текущий перечень полок: ', end='')
        print(*list(directories.keys()), sep=', ')
    elif number_del in directories and directories[number_del] != []:
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', end='')
        print(*list(directories.keys()), sep=', ')
    else:
        print('Такой полки не существует. Текущий перечень полок: ', end='')
        print(*list(directories.keys()), sep=', ')
        
print('Список доступных команд:')
print('p - узнать владельца документа по его номеру', 's - по номеру документа узнать на какой полке он храниться', \
      'l - вывести полную информацию по всем документам', 'as - добавить новую папку', 'ds - удалить существующую папку \
      при условии, что она пустая', 'help - список команд', 'q - выход', sep='\n')

command = ''
while command != 'q':
    command = input('Введите команду: ')
    if command == 'p':
        name_by_doc()
    elif command == 's':
        dir_by_doc()
    elif command == 'l':
        all_doc()
    elif command == 'as':
        add_dir()
    elif command == 'ds':
        del_dir()
    elif command == 'help':
        print('Список доступных команд:')
        print('p - узнать владельца документа по его номеру', 's - по номеру документа узнать на какой полке он храниться', \
              'l - вывести полную информацию по всем документам', 'as - добавить новую папку', \
              'ds - удалить существующую полку при условии, что она пустая', 'help - список команд', 'q - выход', sep='\n')
        
````
